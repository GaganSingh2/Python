```python
from tkinter import *

def click(event):
    global scvalue
    text = event.widget.cget("text")
    print(text)
    
    if text == "=":
        if scvalue.get().isdigit():
            value = int(scvalue.get())
            
            
        else:
            value = eval(screen.get())
           
            scvalue.set(value)
            screen.update()
            
            
    
    elif text == "C":
        
        scvalue.set("")
        screen.update()
        
    
    else:
        scvalue.set(scvalue.get() + text)
        screen.update()
    
root = Tk()
root.geometry("744x944")
root.title("Calculator By Bipulpandey")


scvalue = StringVar()
scvalue.set("")
screen = Entry(root, textvar=scvalue, font="lucida 40 bold")
screen.pack(fill=X, ipadx=8, pady=10, padx=10)

frame = Frame(root, bg="Green")
b = Button(frame, text="9", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=5)
b.bind("<Button-1>", click)

b = Button(frame, text="8", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=18, pady=12) 
b.bind("<Button-1>", click)

b = Button(frame, text="7", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=18, pady=12)
b.bind("<Button-1>", click)

frame.pack()

frame = Frame(root, bg="Green")
b = Button(frame, text="6", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=5)
b.bind("<Button-1>", click)

b = Button(frame, text="5", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=18, pady=12) 
b.bind("<Button-1>", click)

b = Button(frame, text="4", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=18, pady=12)
b.bind("<Button-1>", click)

frame.pack()

frame = Frame(root, bg="Green")
b = Button(frame, text="3", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=5)
b.bind("<Button-1>", click)

b = Button(frame, text="2", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=18, pady=12) 
b.bind("<Button-1>", click)

b = Button(frame, text="1", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=18, pady=12)
b.bind("<Button-1>", click)
frame.pack()

frame = Frame(root, bg="Green")
b = Button(frame, text="%", padx=8, pady=8, font="lucida 20 bold")
b.pack(side=LEFT, padx=13, pady=10)
b.bind("<Button-1>", click)

b = Button(frame, text="0", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=18, pady=12) 
b.bind("<Button-1>", click)

b = Button(frame, text=".", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=18, pady=12)
b.bind("<Button-1>", click)
frame.pack()

frame = Frame(root, bg="Green")
b = Button(frame, text="*", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=17, pady=17)
b.bind("<Button-1>", click)

b = Button(frame, text="/", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=18, pady=12) 
b.bind("<Button-1>", click)


b = Button(frame, text="+", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=18, pady=12)
b.bind("<Button-1>", click)
frame.pack()

frame = Frame(root, bg="Green")
b = Button(frame, text="-", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=5)
b.bind("<Button-1>", click)

b = Button(frame, text="=", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=18, pady=12) 
b.bind("<Button-1>", click)

b = Button(frame, text="C", padx=10, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=18, pady=12)
b.bind("<Button-1>", click)

frame.pack()

root.mainloop()

```
